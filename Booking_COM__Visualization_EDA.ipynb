```python
import pandas as pd
df=pd.read_csv('Booking_com_Cleaned_Data.csv')
```

```python
df.drop(columns=['Unnamed: 0'],inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
# Descriptive statistics for numerical features
numerical_summary = df[['Original Price', 'Current Price', 'Taxes and Charges', 'Discount', 'Price per Day', 'Hotel_Rating', 'Reviews_available']].describe()
numerical_summary
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set style
sns.set_style("whitegrid")

# 1. Distribution of Hotel Ratings
plt.figure(figsize=(8, 5))
sns.histplot(df['Hotel_Rating'], bins=20, kde=True, color='blue')
plt.title("Distribution of Hotel Ratings")
plt.xlabel("Hotel Rating")
plt.ylabel("Count")
plt.show()
```

# Hotel Rating:
- The distribution is right-skewed, with most hotels having ratings between 7 and 9.5.
- The mean rating is 7.48, with a maximum rating of 9.9.

```python
# 2. Boxplot for Price per Day
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Price per Day'], color='orange')
plt.title("Boxplot of Price per Day")
plt.xlabel("Price per Day")
plt.show()
```

# Price per Day:
- The boxplot shows that prices vary significantly, with some extreme high values.
- The median daily price is around 4149.61, and the maximum is 30,322.

```python
# 3. Bar plot for Rating Categories
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Rating_Category', palette='viridis', order=df['Rating_Category'].value_counts().index)
plt.title("Count of Hotels by Rating Category")
plt.xlabel("Rating Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

# Original Price vs. Current Price:
- Many hotels have an original price of 0, likely due to No Discounts.
- The mean current price is 143,805, but with a large standard deviation, indicating high variance in pricing.

# Taxes and Charges:
- The mean tax is 24,008, but some hotels have taxes exceeding 322,056.



# Categorical Data:

# Rating Categories:
- The majority of hotels fall into "Good" and "Fabulous" categories.
# Preferred Hotels:
- A large proportion of hotels are marked as "Preferred."
# Free Cancellation & Pay at Property:
- Many hotels provide free cancellation options.

```python
# 4. Scatter plot: Hotel Rating vs. Price per Day
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Hotel_Rating'], y=df['Price per Day'], alpha=0.6, color='purple',marker='x')
plt.title("Hotel Rating vs. Price per Day")
plt.xlabel("Hotel Rating")
plt.ylabel("Price per Day")
plt.show()
```

# Hotel Rating vs. Price per Day:

- No strong correlation between hotel rating and price per day.
- Some highly rated hotels have low prices, and some lower-rated ones have high prices.

```python
# 5. Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df[['Original Price', 'Current Price', 'Taxes and Charges', 'Discount', 'Price per Day', 'Hotel_Rating', 'Reviews_available']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

# Correlation Heatmap:

- Strong correlation between "Original Price" and "Current Price" (0.10).
- "Price per Day" shows a weak correlation with "Hotel Rating" (0.08), indicating price alone doesn’t determine hotel rating.
- "Reviews Available" has a weak correlation with "Hotel Rating" (0.23), meaning reviews don’t significantly impact ratings.

```python
# 6. Boxplot of Price per Day across Rating Categories
plt.figure(figsize=(8, 5))
sns.boxplot(x='Rating_Category', y='Price per Day', data=df, palette='coolwarm', order=df['Rating_Category'].value_counts().index)
plt.title("Price per Day across Rating Categories")
plt.xlabel("Rating Category")
plt.ylabel("Price per Day")
plt.xticks(rotation=45)
plt.show()
```

# Price per Day across Rating Categories:

- "Superb" and "Fabulous" hotels tend to have higher median prices.
- "Good" category has a wider price range, suggesting varied pricing strategies.

```python
# Crosstab analysis: Free cancellation vs. Rating Category
cancellation_vs_rating = pd.crosstab(df['Free_cancellation'], df['Rating_Category'], normalize='index') * 100
cancellation_vs_rating
```

```python
# Pivot Table: Average Price per Day by Destination and Rating Category
pivot_price_rating = df.pivot_table(values='Price per Day', index='Destination', columns='Rating_Category', aggfunc='mean')
pivot_price_rating
```

```python
# Groupby analysis: Average price per day by rating category
groupby_price_rating = df.groupby('Rating_Category')['Price per Day'].mean().sort_values(ascending=False)
groupby_price_rating
```

# Free Cancellation vs. Rating Category

- Hotels with free cancellation are more common in "Very Good" (22.16%) and "Good" (21.63%) categories.
- "Superb" and "Fabulous" hotels also have free cancellation but at lower percentages.
- Hotels without free cancellation tend to be in the "Good" and "Very Good" categories.

# Average Price per Day by Destination & Rating Category

- New Delhi has the highest-priced "Exceptional" hotels (₹10,597 per day).
- Goa has some of the most expensive "Superb" hotels (₹8,812 per day).
- Kerala has the highest-priced "Not Available" category hotels (₹14,959 per day).
- Jaipur has relatively cheaper hotels in all categories.

# Average Price per Day by Rating Category

- "Superb" hotels have the highest average price per day (₹7,005), followed by "Fabulous" (₹6,930).
- "Good" hotels have the lowest average price per day (₹5,057).
- "Review Score" category hotels (₹4,824) are cheaper compared to higher-rated categories.


# Conclusion:
- Higher-rated hotels generally have higher prices, but location plays a crucial role.
- Free cancellation is more common in "Good" and "Very Good" hotels.
- Goa and New Delhi have some of the priciest hotels in the "Exceptional" and "Superb" categories.
