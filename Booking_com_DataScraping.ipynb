```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
import numpy as np

# Initialize an empty DataFrame to store all the data
df = pd.DataFrame()

# List of cities to search for
list_city = ['Goa', 'Jaipur', 'Rishīkesh', 'New+Delhi', 'Kerala','Bangalore']

for k in list_city:
    driver = webdriver.Chrome()
    driver.get(f'https://www.booking.com/searchresults.en-gb.html?ss={k}')
    # Let the JavaScript execute
    driver.implicitly_wait(20)  # wait for 20 seconds
    
    driver.maximize_window()
    
    try:
        # Wait for the button to appear and click it
        button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//button[@aria-label="Dismiss sign in information."]'))
        )
        button.click()
    except TimeoutException:
        print("Element not found or didn't appear within the wait time")

    driver.implicitly_wait(30)  # wait for 30 seconds

    try:
         # Wait for the button to appear and click it
       
        button = WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.XPATH, "//span[text()=\"I'm flexible\"]"))
        )
        button.click()
    except TimeoutException:
        print("Element not found or didn't appear within the wait time")
        

    #try:
        # Wait until the button is available and then click it
        #button = WebDriverWait(driver, 10).until(
        #EC.element_to_be_clickable((By.XPATH, '//*[@data-testid="searchbox-dates-container"]'))
        #)
        #button.click()
    #except Exception as e:
        #print(f"An error occurred: {e}")

    
        
    try:
        # Wait for the button to appear and click it
        button1 = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//div[text()=\"A month\"]"))
        )
        button1.click()
    except TimeoutException:
        print("Element not found or didn't appear within the wait time")
    
    try:
        button1 = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//span[text()=\"May\"]"))
        )
        button1.click()
    except TimeoutException:
        print("Element not found or didn't appear within the wait time")
    
    try:
        button1 = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//span[text()=\"Select dates\"]"))
        )
        button1.click()
    except TimeoutException:
        print("Element not found or didn't appear within the wait time")

    try:
        search_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//span[text()=\"Search\"]"))
        )
        search_button.click()
    except TimeoutException:
        print("Element not found or didn't appear within the wait time")


    driver.implicitly_wait(20)  # wait for 20 seconds
    
    try:
        # Scroll down to the end of the page
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        # Wait for a few seconds to let the content load
        time.sleep(30)
    except Exception as e:
        print(f"An error occurred during click {counter}: {e}")

    try:
        search_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//span[text()=\"Load more results\"]"))
        )
        search_button.click()
    except TimeoutException:
        print("Element not found or didn't appear within the wait time")
    
    WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, '[data-testid="recommended-units"]'))
    )

        
    # Get the HTML content
    html = driver.page_source
    
    # Parse the content with BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')
    
    # Close the WebDriver
    driver.quit()
    
    hotel_name = []
    hotel_features=[]
    hotel_location=[]
    hotel_price=[]
    hotel_rating=[]
    hotel_Preferred=[]
    hotel_star_Rating=[]

    for i in soup.find_all("div", attrs={"data-testid": "title"}):
        text = i.text.strip()
        if len(text) == 0:
            hotel_name.append(np.nan)
        else:
            hotel_name.append(text)

    for j in soup.find_all("div", attrs={"data-testid": "recommended-units"}):
        text = j.text.strip()
        if len(text) == 0:
            hotel_features.append(np.nan)
        else:
            hotel_features.append(text)

    for i in soup.find_all("span", attrs={"data-testid": "address"}):
        text = i.text.strip()
        if len(text) == 0:
            hotel_location.append(np.nan)
        else:
            hotel_location.append(text)

    for i in soup.find_all("div", attrs={"data-testid": "review-score"}):
        text = i.text.strip()
        if len(text) == 0:
            hotel_rating.append(np.nan)
        else:
            hotel_rating.append(text)
    
    for i in soup.find_all("div", attrs={"data-testid": "availability-rate-information"}):
        text = i.text.strip()
        if len(text) == 0:
            hotel_price.append(np.nan)
        else:
            hotel_price.append(text)



    for i in soup.find_all("span", class_="fcd9eec8fb c2a6770498 b3d142134a c2cc050fb8 e410954d4b"):
        text = i.get('aria-label')
        if len(text) == 0:
            hotel_Preferred.append(np.nan)
        else:
            hotel_Preferred.append("YES")

        hotel_star_Rating = []
    

    for i in soup.find_all("div", class_="b3f3c831be"):
        text = i.get('aria-label')
        if len(text) == 0:
            hotel_star_Rating.append(np.nan)
        else:
            hotel_star_Rating.append(text)
    
         
    print("Location= ",k)
    print(len(hotel_name))
    print(len(hotel_features))
    print(len(hotel_location))
    print(len(hotel_rating))
    print(len(hotel_price))
    print(len(hotel_Preferred))
    print(len(hotel_star_Rating))

    # Ensure all lists are of the same length by padding shorter lists with NaN or a default value
    max_length = max(len(hotel_name), len(hotel_features), len(hotel_location), len(hotel_rating), len(hotel_price), len(hotel_Preferred), len(hotel_star_Rating))

    hotel_name += [np.nan] * (max_length - len(hotel_name))
    hotel_features += [np.nan] * (max_length - len(hotel_features))
    hotel_location += [np.nan] * (max_length - len(hotel_location))
    hotel_rating += [np.nan] * (max_length - len(hotel_rating))
    hotel_price += [np.nan] * (max_length - len(hotel_price))
    hotel_Preferred += [np.nan] * (max_length - len(hotel_Preferred))
    hotel_star_Rating += [0] * (max_length - len(hotel_star_Rating))  # Using 0 for hotel_star_Rating
    
    # Create a DataFrame for the current city
    df1 = pd.DataFrame({
        "Hotel_Name": hotel_name,
        "Facilities": hotel_features,
        "Location": hotel_location,
        "Rating": hotel_rating,
        "Price": hotel_price,
        "Destination": k,
        "Preferred_Hotel":hotel_Preferred,
        "Hotel_Star_Rating":hotel_star_Rating
    })
    
    # Append the DataFrame to the main DataFrame
    df = pd.concat([df, df1], ignore_index=True)

#print(df)
df.to_csv('Booking_com_Cleaned_Data.csv')
```
