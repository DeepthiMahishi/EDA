```python
# Function to calculate discount
def calculate_discount(row):
    if pd.isna(row['Original Price']):
        return 0
    else:
        return row['Original Price'] - row['Current Price']

# Apply the function to calculate the discount
df['Discount'] = df.apply(calculate_discount, axis=1)
```

```python
# Function to calculate Total amount
def calculate_Total(row):
    if pd.isna(row['Taxes and Charges']):
        return 0 + row['Current Price']
    else:
        return row['Taxes and Charges'] + row['Current Price']

# Apply the function to calculate the discount
df['Total Amount Payable'] = df.apply(calculate_Total, axis=1)
```
